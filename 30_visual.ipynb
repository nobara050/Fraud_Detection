## Prepare environment


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Run this cell only if you want a IEEE style graph.

## Load data

```python
results = pd.read_csv('./output/results.txt')
results.head()
```

```python
results.columns
```

## Visualization

```python
plt.plot(results['Epoch'],results['Loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.savefig('./output/loss.jpg')
plt.show()
```

```python
plt.plot(results['Epoch'],results['F1'])
plt.xlabel('Epoch')
plt.ylabel('F1 score')
plt.savefig('./output/f1.jpg')
plt.show()
```

```python
total_min = np.sum(results['Time(s)']) / 60
print('Time cost:', round(total_min,2), 'minutes.')
```

```python
# y-axis in log scale
plt.plot(results['Epoch'],results['Loss'])
plt.yscale('log')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
